Write `safe_divide(a, b)` that:
- Returns `a / b` normally
- Returns `None` if `b == 0` (catch `ZeroDivisionError`)
- Raises `TypeError` with message `"Inputs must be numeric"` if either input is not a number

````python
def safe_divide(a, b):
    # YOUR CODE HERE
    if not isinstance(a, (int, float)) or not isinstance(b, (int, float)):
        raise TypeError("Inputs must be numeric")
    try:
        return a / b
    except ZeroDivisionError:
        return None

# Tests
assert safe_divide(10, 2) == 5.0
assert safe_divide(5, 0) is None
try:
    safe_divide('x', 2)
    assert False, "Should have raised TypeError"
except TypeError as e:
    assert str(e) == "Inputs must be numeric"
print("1.3 passed ✓")
````

**Your Geometric Explanation (4.3):**

### Geometric Explanation

A matrix transformation can change the size and direction of vectors. However, some vectors have a special property: after the transformation, they still point in the same direction (or exactly opposite). These special vectors are called eigenvectors.

The corresponding eigenvalue tells us how much the vector is stretched or compressed. If the eigenvalue is greater than 1, the vector becomes longer. If it is between 0 and 1, the vector becomes shorter. A negative eigenvalue means the vector flips direction while also changing its length.

In simple terms, eigenvectors show the directions that are preserved by a transformation, and eigenvalues describe how much those directions are scaled. This geometric interpretation helps us understand how matrices affect shapes and spaces and is useful in areas such as computer graphics, image processing, and machine learning.

### 4.4 SVD & Dimensionality Reduction

1. Create a random 4×3 matrix `X` (use `np.random.seed(42)`).
2. Compute its **SVD**: `U, S, Vt = np.linalg.svd(X, full_matrices=False)`.
3. **Reconstruct** `X` from the SVD and verify it matches the original.
4. Perform a **rank-1 approximation** using only the largest singular value/vector — call it `X_approx`.
5. Explain the relationship between SVD components and PCA in the markdown cell below.

````python
np.random.seed(42)
X = np.random.randn(4, 3)

U, S, Vt = np.linalg.svd(X, full_matrices=False)  # YOUR CODE HERE

X_reconstructed = U @ np.diag(S) @ Vt  # YOUR CODE HERE — reconstruct from U, S, Vt

X_approx = S[0] * np.outer(U[:, 0], Vt[0, :])  # YOUR CODE HERE — rank-1 approximation

print(f"U shape: {U.shape}, S shape: {S.shape}, Vt shape: {Vt.shape}")
print(f"Singular values: {S}")
print(f"Reconstruction error: {np.linalg.norm(X - X_reconstructed):.2e}")
print(f"Rank-1 approximation:\n{X_approx}")

assert np.allclose(X, X_reconstructed, atol=1e-10), "Reconstruction failed!"
print("4.4 passed ✓")
````

**SVD → PCA Connection (4.4):**

> *Explain: what are U, S, Vt in SVD, and how do they relate to principal components? Which matrix contains the directions of maximum variance?*

### SVD → PCA Connection

SVD breaks a matrix into three parts: **U, S, and Vt**. Each part has a different role in describing the data.

- **U** tells us how the data points are arranged along important patterns.
- **S** contains the singular values, which show how much information or variation is present in each pattern.
- **Vt** contains the directions along which the data varies the most.

These directions in **Vt** are the same as the principal components used in PCA. In other words, **Vt contains the directions of maximum variance**, while the values in **S** tell us how important each direction is.

For example, imagine a dataset containing students' marks in Math, Science, and English. The marks in these subjects are often related. PCA uses the directions stored in **Vt** to find a new set of axes that capture most of the differences among students. Instead of analyzing three separate subjects, we can summarize the data using one or two principal components without losing much information.

So, SVD provides the mathematical foundation for PCA: **Vt gives the principal directions, and S tells us how much variance each direction explains.**
